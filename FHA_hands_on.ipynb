```python
import random
import pandas as pd
from collections import Counter


# Step 1: Generator, it Asks for a word and Generates a Shakespearean-style sentence


def shakespeare_generator(word, extra_words=None):
    openings = ["Lo,", "Hark!", "Behold,", "Alas,", "Verily,"]

    base_middles = [
        f"the spirit of {word} doth wander through the silent night",
        f"{word} blooms fair beneath the pale moon",
        f"in thy {word} lies the seed of destiny",
        f"{word} art a tempest in the hearts of men",
        f"from {word} springs both sorrow and delight"
    ]

    # If we discovered common words, use them
    if extra_words:
        for w in extra_words:
            base_middles.append(f"{w} entwines with {word} in solemn reverie")

    endings = [
        "thus fate unfolds.",
        "so speaketh the stars.",
        "and time bears witness.",
        "till the morrow comes.",
        "as heaven decrees."
    ]

    return f"{random.choice(openings)} {random.choice(base_middles)}, {random.choice(endings)}"



# Step 2: Initial Generation


word = input("Enter a word, good scholar: ")

sentences = [shakespeare_generator(word) for _ in range(10)]
df = pd.DataFrame({"sentence": sentences})


# Step 3: Analysis, Analyze word frequencies


all_words = " ".join(df["sentence"]).lower().replace(",", "").replace(".", "").split()
word_freq = Counter(all_words)

print("\nðŸ“Š Most Common Words:")
common_words = []

for w, c in word_freq.most_common(5):
    print(w, ":", c)
    common_words.append(w)


# Step 4: Feedback Loop, using the most common words to bias future generation


print("\n Regenerating using learned patterns...\n")

new_sentences = [shakespeare_generator(word, extra_words=common_words) for _ in range(5)]

for s in new_sentences:
    print(s)
```
