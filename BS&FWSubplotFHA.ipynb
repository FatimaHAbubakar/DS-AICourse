```python
#Subplot of Bathing Soap and Face wash
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("company_sales_data.csv")

months = df["month_number"]
bathing_soap = df["bathingsoap"]
facewash = df["facewash"]

# Create subplots: 1 row, 2 columns
plt.figure(figsize=(10,4))

# First subplot  Bathing soap
plt.subplot(1, 2, 1)
plt.plot(months, bathing_soap, color='blue', marker='o')
plt.title("Bathing Soap Sales")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.xticks(months)

# Second subplot  Facewash
plt.subplot(1, 2, 2)
plt.plot(months, facewash, color='purple', marker='s')
plt.title("Facewash Sales")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.xticks(months)

plt.tight_layout()  # makes sure plots dont overlap
plt.show()
```
