```python
#Analyze a small, publicly available dataset using Python's NumPy or Pandas libraries.

import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target


# PART 1: EXPLORE THE DATA


print("FIRST FIVE ROWS:")
print(data.head())

print("\nDATA INFORMATION:")
print(data.info())

print("\nSUMMARY STATISTICS:")
print(data.describe())


# PART 2: BASIC STATISTICS


print("\nMEAN:")
print(data.mean(numeric_only=True))

print("\nMEDIAN:")
print(data.median(numeric_only=True))

print("\nMODE:")
print(data.mode(numeric_only=True))

print("\nSTANDARD DEVIATION:")
print(data.std(numeric_only=True))


# CORRELATION MATRIX


print("\nCORRELATION BETWEEN FEATURES:")
print(data.corr(numeric_only=True))

```
