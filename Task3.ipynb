```python

# Create a basic line plot and scatter plot with Matplotlib.
# Visualize a dataset with histograms and box plots using Seaborn.

# DATA VISUALIZATION WITH MATPLOTLIB & SEABORN


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris


iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target


# MATPLOTLIB VISUALIZATIONS

# Line Plot

plt.figure(figsize=(8, 5))
plt.plot(data.index, data['sepal length (cm)'], label='Sepal Length')
plt.plot(data.index, data['sepal width (cm)'], label='Sepal Width')
plt.title("Line Plot of Sepal Length and Width")
plt.xlabel("Sample Index")
plt.ylabel("Measurement (cm)")
plt.legend()
plt.show()

# Scatter Plot

plt.figure(figsize=(8, 5))
plt.scatter(data['sepal length (cm)'], data['petal length (cm)'], c=data['species'], cmap='viridis')
plt.title("Scatter Plot: Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.colorbar(label='Species')
plt.show()

# SEABORN VISUALIZATIONS


# Histogram

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='petal width (cm)', kde=True, bins=20)
plt.title("Histogram of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.show()

# Box Plot

plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=data)
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species (0=setosa, 1=versicolor, 2=virginica)")
plt.ylabel("Sepal Length (cm)")
plt.show()

```
