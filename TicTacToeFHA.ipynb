```python
import random
#The AI agent learns by storing board states that lead to winning outcomes and reusing those moves when similar states occur in future games.
#The board is represented by a 3x3 matrix, inputing a number places the current player X on that position.


# Game board
board = [" " for _ in range(9)]

# Memory for AI learning
ai_memory = {}

# Display the board
def print_board():
    print()
    for i in range(0, 9, 3):
        print(board[i], "|", board[i+1], "|", board[i+2])
    print()

# Check winner
def check_winner(player):
    win_conditions = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

# Check tie
def check_tie():
    return " " not in board

# Get available moves
def available_moves():
    return [i for i in range(9) if board[i] == " "]

# AI move (learns from past wins)
def ai_move():
    state = tuple(board)
    if state in ai_memory:
        return ai_memory[state]
    return random.choice(available_moves())

# Main game loop
def play_game():
    current_player = "X"

    while True:
        print_board()

        if current_player == "X":
            move = int(input("Choose position (0-8): "))
            if board[move] != " ":
                print("Invalid move.")
                continue
        else:
            move = ai_move()
            print(f"AI chooses position {move}")

        board[move] = current_player

        if check_winner(current_player):
            print_board()
            print(f"{current_player} wins!")

            # AI learns winning move
            if current_player == "O":
                ai_memory[tuple(board)] = move
            break

        if check_tie():
            print_board()
            print("It's a tie!")
            break

        current_player = "O" if current_player == "X" else "X"

# Call the main game loop
play_game()
```
