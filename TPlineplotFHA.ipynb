```python
#Line plot of Total Profit of all months
import pandas as pd
import matplotlib.pyplot as plt

# load csv
df = pd.read_csv("company_sales_data.csv")

# get months and total profit
months = df["month_number"]
total_profit = df["total_profit"]

# Plot line
plt.plot(months, total_profit, marker='o', color='orange')
plt.title("Total Profit of All Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
 # show month numbers properly
plt.xticks(months)
plt.grid(True)
plt.show()


```
